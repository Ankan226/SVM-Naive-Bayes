**24.  Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary:**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.svm import SVC

iris = load_iris()
X = iris.data[:, :2]  # use only 2 features for plotting
y = iris.target

model = SVC(kernel='poly', degree=3)
model.fit(X, y)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=y)

# Meshgrid
import numpy as np
x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                     np.linspace(y_min, y_max, 200))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Polynomial Kernel SVM Decision Boundary")
plt.show()
```

**26.  Write a Python program to train a Multinomial Nave Bayes classifier for text classification using the 20

**29. Write a Python program to apply feature scaling before training an SVM model and compare results with
unscaled data.**

```python
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X, y = datasets.load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=42)


m1 = SVC().fit(X_train, y_train)


sc = StandardScaler()
X_train_s = sc.fit_transform(X_train)
X_test_s = sc.transform(X_test)
m2 = SVC().fit(X_train_s, y_train)

print("Unscaled:", accuracy_score(y_test, m1.predict(X_test)))
print("Scaled:", accuracy_score(y_test, m2.predict(X_test_s)))
```

**40. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and
compare accuracy.**

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVC
from sklearn import datasets
from sklearn.model_selection import train_test_split

X,y = datasets.load_iris(return_X_y=True)
X_train,X_test,y_train,y_test = train_test_split(X,y)

rfe = RFE(SVC(kernel="linear"), n_features_to_select=2).fit(X_train,y_train)
m = SVC().fit(rfe.transform(X_train), y_train)

print(m.score(rfe.transform(X_test), y_test))
```
